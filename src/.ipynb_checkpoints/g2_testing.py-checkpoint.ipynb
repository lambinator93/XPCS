g_2(\mathbf{Q},t) = \frac{\langle I(\mathbf{Q},t)I(\mathbf{Q},t+\tau)\rangle_{\tau}}{\langle I(\mathbf{Q},\tau)\rangle^2_{\tau}}
$$

# Compare getdelayinfo Results

```python
numFrames = np.arange(1,10000+1,1)

maxLevels = np.zeros((1,len(numFrames)))

for i in range(len(numFrames)):
    maxLevels[0,i]+=getdelayinfo(numFrames[i],4)[1]
    
maxLevelsMat = np.array(scipy.io.loadmat('maxLevels_1_to_10000.mat')['maxLevels'])


import matplotlib.pyplot as plt
%matplotlib notebook

plt.figure
plt.plot(numFrames,maxLevels[0,:],'b--')
plt.plot(numFrames,maxLevelsMat[0,:],'r')
plt.plot(numFrames,maxLevels[0,:]-maxLevelsMat[0,:],'g')
plt.legend(('python','matlab','error'))
plt.ylabel('maxLevel')
plt.xlabel('Number of Frames')
plt.title('Comparison of getdelayinfo Functions for dpl=4')
```

```python
F = open('test_speckle.npy','rb')
IMG = np.load(F)
MASK = np.zeros((IMG.shape[1:]))


roi_size = 50
MASK[0:roi_size,0:roi_size] = 1
```

```python
sumI,G2,_,_,_ = g2calc_parallel(IMG,MASK)

G2 = np.mean(G2,axis=(1,2))
```

```python
G2Mat = np.array(scipy.io.loadmat('G2_matlab.mat')['G2'])
G2Mat = np.mean(G2Mat,axis=(1,2))
```

```python
framespacing = 1 # second
nframes = IMG.shape[0]
dpl = 4
delay = framespacing * finddelays(nframes,dpl,1)
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
plt.figure
notzero = 1*10**(-12)
plt.semilogx(delay[0,:],G2[:-1]/np.max(G2[:-1]),'b*')
plt.semilogx(delay[0,:],G2Mat[:-1]/np.max(G2Mat[:-1]),'r.')
plt.semilogx(delay[0,:],np.abs(G2[:-1]-G2Mat[:-1])/G2[:-1],'g')

plt.legend(('python','matlab','error'))
plt.ylabel('g2')
plt.xlabel('time delays (s)')
plt.title('Comparison of g2 Functions on Simulated Data for dpl=4')
```
